# Challenge 2 - Data Cleaning and Transformation

```python
# Your code here
def clean_outliers(df, columns, method="iqr", threshold=1.5):
    """
    Removes outliers from the specified columns in a DataFrame using the IQR method.
    
    Args:
        df (pd.DataFrame): The DataFrame to process.
        columns (list): List of column names to clean.
        method (str): Method to detect outliers ('iqr' or 'zscore').
        threshold (float): Threshold for outlier detection. Default is 1.5 for IQR.
    
    Returns:
        pd.DataFrame: A new DataFrame with outliers removed.
    """
    cleaned_df = df.copy()
    
    for col in columns:
        if method == "iqr":
            # Calculate Q1 (25th percentile) and Q3 (75th percentile)
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Determine outlier boundaries
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            
            # Filter out outliers
            cleaned_df = cleaned_df[(cleaned_df[col] >= lower_bound) & (cleaned_df[col] <= upper_bound)]
        
        elif method == "zscore":
            # Z-score method
            mean = df[col].mean()
            std = df[col].std()
            cleaned_df = cleaned_df[(df[col] >= mean - threshold * std) & (df[col] <= mean + threshold * std)]
    
    return cleaned_df

# Columns to clean
columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Clean the DataFrame using IQR method
cleaned_df = clean_outliers(customers_df, columns, method="iqr", threshold=1.5)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
def plot_elbow_method(data, max_clusters=10):
    """
    Plot the Elbow Method to find the optimal number of clusters for K-means.

    Args:
    data (array-like): Scaled numerical data for clustering.
    max_clusters (int): Maximum number of clusters to test (default=10).
    """
    wcss = []  # Store WCSS values for each number of clusters
    
    # Fit K-Means with different numbers of clusters
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)  # Inertia is the WCSS

    # Plot the Elbow Curve
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method for Optimal Clusters')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
    plt.xticks(range(1, max_clusters + 1))
    plt.grid()
    plt.show()

# Example Usage
# Assuming 'df_scaled' is the scaled version of your data
plot_elbow_method(customers_scale, max_clusters=10)
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2, random_state=42).fit(customers_scale)

cleaned_df['Label'] = kmeans_2.labels_.tolist()

cleaned_df['Label'].value_counts()
```

```python
# Define cluster labels and colors for 2 clusters
cluster_labels = ['Cluster 1', 'Cluster 2']
colors = ['#1f77b4', '#ff7f0e']  # Blue and orange

# Create a figure
plt.figure(figsize=(10, 8))

# Loop through clusters to plot each with its own color
for cluster_idx in range(2):
    plt.scatter(customers_scale[np.array(clusters) == cluster_idx, 0],  # Filter rows by cluster label
                customers_scale[np.array(clusters) == cluster_idx, 1], 
                color=colors[cluster_idx], 
                label=cluster_labels[cluster_idx], 
                s=100, 
                edgecolor='k', 
                alpha=0.8)

# Plot centroids
plt.scatter(kmeans_2.cluster_centers_[:, 0], kmeans_2.cluster_centers_[:, 1], 
            s=300, c='red', marker='X', label='Centroids', edgecolor='black')

# Add labels, title, and legend
plt.title('K-Means Clusters (2 Clusters)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Feature 1', fontsize=14, labelpad=10)
plt.ylabel('Feature 2', fontsize=14, labelpad=10)
plt.legend(loc='upper right', fontsize=12, frameon=True)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

# Your code here
model = DBSCAN(eps=0.5).fit(customers_scale)

cleaned_df['labels_DBSCAN'] = model.labels_.tolist()
```

Count the values in `labels_DBSCAN`.

```python
# Your code here
cleaned_df['labels_DBSCAN'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define cluster labels
clusters = cleaned_df['labels_DBSCAN'].values  # Extract DBSCAN cluster labels
unique_labels = set(clusters)  # Unique labels from DBSCAN

# Define colors: contrasting colors for clusters and black for noise
colors = ['#1f77b4', '#ff7f0e']  # Blue and orange for clusters
noise_color = 'red'  # Black for noise points

# Create a figure
plt.figure(figsize=(10, 8))

# Loop through each cluster label and plot
for label in unique_labels:
    cluster_points = clusters == label  # Boolean mask for the current cluster
    if label == -1:  # Noise points
        plt.scatter(
            customers_scale[cluster_points, 0],  # X-axis feature
            customers_scale[cluster_points, 1],  # Y-axis feature
            c=noise_color, 
            label='Noise', 
            s=50, 
            edgecolor='k', 
            alpha=0.6
        )
    else:  # Regular clusters
        plt.scatter(
            customers_scale[cluster_points, 0],  # X-axis feature
            customers_scale[cluster_points, 1],  # Y-axis feature
            c=colors[label % len(colors)],  # Cycle through contrasting colors
            label=f'Cluster {label}', 
            s=100, 
            edgecolor='k', 
            alpha=0.8
        )

# Add labels, legend, and title
plt.title('DBSCAN Clustering', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Feature 1', fontsize=14, labelpad=10)  # Replace with actual feature name
plt.ylabel('Feature 2', fontsize=14, labelpad=10)  # Replace with actual feature name
plt.legend(loc='upper right', fontsize=12, frameon=True)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

Noise Handling:

DBSCAN identifies some data points as noise (-1 cluster), which K-Means cannot do because it assigns every data point to a cluster. This shows that DBSCAN is better at identifying outliers or anomalies in the dataset.

**Cluster Characteristics:**

- K-Means Clusters: Groups are more generalized and may not align well with actual data distributions, as K-Means assumes spherical clusters.
- DBSCAN Clusters: Identifies a more nuanced structure by differentiating clusters based on density.
For example:
        labels_DBSCAN = 0: Lower values for Milk, Grocery, and Detergents_Paper.
        labels_DBSCAN = 1: Moderately higher values for Fresh and Grocery.
        labels_DBSCAN = -1: Represents potential outliers, as these rows have significantly different characteristics.

**Suitability for Data:**

- DBSCAN performs better here as the clusters are more specific and handle noise effectively.
- K-Means may oversimplify the clustering due to its reliance on centroids and predefined cluster counts.

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:
kmeans_3 = KMeans(n_clusters=3, random_state=42).fit(customers_scale)

cleaned_df['label_kmeans_3'] = kmeans_3.labels_.tolist()

cleaned_df['label_kmeans_3'].value_counts()
```

```python
# Define cluster labels and colors for 3 clusters
cluster_labels = ['Cluster 1', 'Cluster 2', 'Cluster 3']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, orange, green

# Create a figure
plt.figure(figsize=(10, 8))

# Loop through clusters to plot each with its own color
for cluster_idx in range(3):
    cluster_points = np.array(clusters) == cluster_idx  # Boolean mask for current cluster
    plt.scatter(
        customers_scale[cluster_points, 0],  # X-axis feature
        customers_scale[cluster_points, 1],  # Y-axis feature
        color=colors[cluster_idx], 
        label=cluster_labels[cluster_idx], 
        s=100, 
        edgecolor='k', 
        alpha=0.8
    )

# Plot centroids
plt.scatter(
    kmeans_3.cluster_centers_[:, 0], 
    kmeans_3.cluster_centers_[:, 1], 
    s=300, c='red', marker='X', label='Centroids', edgecolor='black'
)

# Add labels, title, and legend
plt.title('K-Means Clusters (3 Clusters)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Feature 1', fontsize=14, labelpad=10)  
plt.ylabel('Feature 2', fontsize=14, labelpad=10)  
plt.legend(loc='upper right', fontsize=12, frameon=True)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

**Your comment here**

Based on the plots, the 3-cluster solution seems better at capturing distinct subgroups, especially if the dataset has enough variation and distinct characteristics. However, the final choice should depend on the context of the analysis and the specific use case:
- 3 Clusters: Use when finer distinctions are required, and the green cluster is meaningful.
- 2 Clusters: Use when you need a simpler, broader grouping of data.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here

model = DBSCAN(eps=0.9, min_samples=7).fit(customers_scale)

cleaned_df['labels_DBSCAN_2'] = model.labels_.tolist()

cleaned_df['labels_DBSCAN_2'].value_counts()
```

```python
# Define cluster labels
clusters = cleaned_df['labels_DBSCAN_2'].values  # Extract DBSCAN cluster labels
unique_labels = set(clusters)  # Unique labels from DBSCAN

# Define colors: contrasting colors for clusters and black for noise
colors = ['#1f77b4', '#ff7f0e']  # Blue and orange for clusters
noise_color = 'red'  # Black for noise points

# Create a figure
plt.figure(figsize=(10, 8))

# Loop through each cluster label and plot
for label in unique_labels:
    cluster_points = clusters == label  # Boolean mask for the current cluster
    if label == -1:  # Noise points
        plt.scatter(
            customers_scale[cluster_points, 0],  # X-axis feature
            customers_scale[cluster_points, 1],  # Y-axis feature
            c=noise_color, 
            label='Noise', 
            s=50, 
            edgecolor='k', 
            alpha=0.6
        )
    else:  # Regular clusters
        plt.scatter(
            customers_scale[cluster_points, 0],  # X-axis feature
            customers_scale[cluster_points, 1],  # Y-axis feature
            c=colors[label % len(colors)],  # Cycle through contrasting colors
            label=f'Cluster {label}', 
            s=100, 
            edgecolor='k', 
            alpha=0.8
        )

# Add labels, legend, and title
plt.title('DBSCAN Clustering', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Feature 1', fontsize=14, labelpad=10)  # Replace with actual feature name
plt.ylabel('Feature 2', fontsize=14, labelpad=10)  # Replace with actual feature name
plt.legend(loc='upper right', fontsize=12, frameon=True)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

**Your comment here**

eps=0.9 is better compared to eps=0.5 for the following reasons:

- It reduces the proportion of noise points, making the results more interpretable.
- It identifies one major cluster (Cluster 0) that likely represents a meaningful structure in the data.

eps=0.9 provides a balance between identifying dense clusters and filtering noise, while eps=0.5 is too strict, classifying most points as noise.
